```python
df = pd.read_csv('toronto_ds.csv')
```

```python
df.drop('Unnamed: 0',inplace=True,axis=1)
```

## let's try to explore some neighbourhood 

```python
df.loc[df['Borough'] == 'West Toronto']
```

## i chose some random neighbourhood 

```python
client_id = '************************************'
client_secret = '********************************'
version = '20180323'
place = df.iloc[76].Neighbourhood
latitude = df.iloc[76].Latitude
longitude = df.iloc[76].Longitude
endpoint = 'explore'
```

```python
print(f'lets explore this place : {place}')
```

```python
LIMIT = 100
radius = 500

url = 'https://api.foursquare.com/v2/venues/{}?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    endpoint,
    client_id, 
    client_secret, 
    version, 
    latitude, 
    longitude, 
    radius, 
    LIMIT)
url
```

```python
res = requests.get(url)
```

```python
data = res.json()
data
```

## The above json data looking ugly , lets make it clean 

```python
venues = data['response']['groups']
for i in venues[0]:
    print(i)
```

```python
venues[0]['items'][0].keys()
```

## json_normalise from pandas will make a df out of this json data

```python
vdf = json_normalize(venues[0]['items']) 
```

```python
explore = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
explored_df = vdf.loc[:,explore]
```

```python
explored_df.head()
```

## now's its looking clean except venue.categories

```python
for i in vdf['venue.categories']:
    print(i[0].keys())
```

```python
def clean_category(cat):
    name = cat[0]['name']
    return name
```

```python
get_name = lambda x: clean_category(x)
```

```python
category = explored_df['venue.categories'].apply(get_name)
```

```python
explored_df['venue.categories'] = category
```
